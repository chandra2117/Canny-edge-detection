```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
original_image=cv2.imread("images.jpeg")
```

```python
plt.imshow(original_image[:,:,::-1])
```

```python
blur_image=cv2.blur(original_image, ksize=(3,3))
```

```python
plt.imshow(blur_image[:,:,::-1])
```

```python
medium=np.median(original_image)
```

```python
low=int(max(0,0.7*medium))
high=int(max(255,1/3*medium))
```

```python
canny_edge=cv2.Canny(blur_image, threshold1=low, threshold2=high)
```

```python
plt.imshow(canny_edge,cmap="gray")
```
